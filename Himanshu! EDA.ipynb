**GRAPH PRESENTATION**

```python
df = pd.read_csv('/content/drive/MyDrive/train.csv')
# Create the subset dataframe
subset_df = df.head(50000)
df = subset_df.dtypes # use subset_df
# 10 countries which is having highest number of confirmed cases
df = pd.read_csv('/content/drive/MyDrive/subset_Train.csv')
confirmed_cases = df[df['Target'] == 'ConfirmedCases'].groupby('Country_Region')['TargetValue'].sum()
top_10_countries = confirmed_cases.nlargest(10)
print(top_10_countries)
# Total confirmed cases in top 10 countries
confirmed_cases = df[df['Target'] == 'ConfirmedCases'].groupby('Country_Region')['TargetValue'].sum() # Changed 'data' to 'df'
confirmed_cases.nlargest(10).plot(kind='pie', figsize=(20, 8))
plt.title('Top 10 Countries by Total Confirmed Cases')

plt.show()
```

```python
# data is in a CSV file named "covid_data.csv"
covid_data = pd.read_csv('/content/drive/MyDrive/subset_Train.csv')

sns.barplot(data=covid_data, x="Target", y="TargetValue", hue="Target")
plt.title("COVID-19: Confirmed Cases and Deaths")
plt.xlabel("Type of Occurence")
plt.ylabel("No. of Cases/Deaths")
sns.despine()
plt.show()
```

```python
# 10 countries which is having highest number of fatalities
fatalities_by_country = subset_df[subset_df['Target'] == 'Fatalities'].groupby('Country_Region')['TargetValue'].sum()

# create nlargest on the fatalities_by_country Series, not DataFrame
fatalities_by_country.nlargest(10).plot(kind='pie', autopct='%1.1f%%')

plt.title('Fatalities Distribution by Country (Top 10)')
plt.ylabel('')
plt.show()
```

```python
# Correlation graph
correlation_data = covid_data[['Population', 'Weight', 'TargetValue']]

correlation_matrix = correlation_data.corr()

# Plot map
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu', vmin=-1, vmax=1)
plt.title('Correlation Matrix', fontsize=14)
plt.show()
```

```python
# data is in 'subset_Train.csv',
covid_data = pd.read_csv('/content/drive/MyDrive/subset_Train.csv')
print(covid_data.columns)
datewise_trend = covid_data.groupby(['Date', 'Target'])['TargetValue'].sum().reset_index()

# Pivot the table to 'ConfirmedCases' and 'Fatalities' as columns
datewise_trend = datewise_trend.pivot(index='Date', columns='Target', values='TargetValue').reset_index()
datewise_trend.rename(columns={'ConfirmedCases': 'Confirmed Cases', 'Fatalities': 'Fatalities'}, inplace=True)
sns.scatterplot(data=datewise_trend, x="Confirmed Cases", y="Fatalities")
plt.title("Confirmed Cases vs Fatalities")
plt.xlabel("Confirmed Cases")
plt.ylabel("Fatalities")
sns.despine()

# plotting a trendline
x = datewise_trend["Confirmed Cases"]
y = datewise_trend["Fatalities"]
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), "r--")
plt.show()
```
